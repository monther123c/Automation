```python
import pandas as pd
df=pd.read_csv('daily_report_python.csv')
df
```

save  every table alone  

```python
df1,df2=df[:1],df[1:]
df2.columns=df2.iloc[0]
df2.drop(df2.index[0],inplace=True)
df2
```

make  student  email  an id 


```python
df2['Students ID']=df2['User Email'].str.extract(pat=('(^\d*)'))
df2
```

```python
df2["Total"]=df2['Duration (Minutes)'].astype('int')
```

i need to know how long does this meting take in time so i can know when  suudent are late or apsent 


```python
meeting_duration=int(df1['Duration (Minutes)'][0])
meeting_duration
```

```python
df2=df2.groupby('Students ID',as_index=False).sum(numeric_only=True)
df2= df2[1:]
df2
```

```python
df2['Status']=['Present'if x>=meeting_duration-10 else 'Absent' if x<meeting_duration-20 else 'Late' for x in df2['Total']]
df2['Percentage of attendance']=[f'{round((x/meeting_duration)*100)}%' for x in df2['Total']]
df2
```

```python
meeting_id=df1['Meeting ID'][0]
meeting_id
```

saving data into a file 

```python
data=pd.read_csv('data.csv')
data[meeting_id]=df2['Status']
data['Students ID']=df2['Students ID']
data['Percentage of attendance']=df2['Percentage of attendance']
data.to_csv('data.csv')
data
```
